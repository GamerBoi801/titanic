```python
import pandas as pd

def load_data(filename='train.csv'):
    df = pd.read_csv(filename)
    return df

train = load_data()
test = load_data('test.csv')

train['train_flag'] = 1
test['train_flag'] = 0

data = pd.concat([train, test], ignore_index=True)
```

```python
data.describe()
```

```python
"""Removing the columns we dont want"""
#inplace true modifies the original data
data.drop("Name", axis=1, inplace=True)
data.drop("Ticket", axis=1, inplace=True)
data.drop("Cabin", axis=1, inplace=True)
```

```python
data.drop("PassengerId", axis=1, inplace=True)
```

```python
#filling the missing age vals w/ 
age_median = data['Age'].median()
data['Age'].fillna(age_median, inplace=True)
```

```python
#binning th4 age data since the model better categories than inf num stages
bins = [0, 13, 19, 30, 50, 100]
labels = ['child', 'Teen', 'Young', 'Adult', 'Senior']
data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels)
```

```python
#removing the age column
data.drop("Age", axis=1, inplace=True)
```
